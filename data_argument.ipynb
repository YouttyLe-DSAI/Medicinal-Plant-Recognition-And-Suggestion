```python
# Đường dẫn tới thư mục input và output
input_folder = '/content/BongBao'
output_folder = 'data'
```

```python
# Tạo các phép biến đổi
seq = iaa.Sequential([
    #iaa.Affine(rotate=(-45, 45)), # Xoay ảnh từ -45 đến 45 độ
    #iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)}), # Dịch chuyển ảnh
    iaa.Affine(scale=(0.8, 1.2)), # Phóng to/thu nhỏ
    iaa.Fliplr(0.5), # Lật ảnh ngang
    iaa.Flipud(0.5), # Lật ảnh dọc
    iaa.Multiply((0.8, 1.2)), # Thay đổi độ sáng
    # iaa.AdditiveGaussianNoise(scale=(10, 30)), # Thêm nhiễu Gaussian
    iaa.Crop(percent=(0, 0.2)), # Cắt ngẫu nhiên từ 0 đến 20%
    iaa.GaussianBlur(sigma=(0, 3.0)), # Làm mờ ảnh với độ lệch chuẩn Gaussian từ 0 đến 3.0
    # iaa.LinearContrast((0.75, 1.5)), # Thay đổi độ tương phản từ 0.75 đến 1.5 lần
])
```

Chạy 1 trong 2

```python
# Tạo thư mục 'data' nếu chưa tồn tại
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Lấy danh sách tất cả các file ảnh trong thư mục input
image_files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

# Duyệt qua từng file ảnh trong thư mục input
for image_file in image_files:
    # Đọc ảnh
    image_path = os.path.join(input_folder, image_file)

    image = cv2.imread(image_path)
    if image is not None:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Tạo các ảnh đã biến đổi
        images_aug = seq.augment_images([image] * 2)

    # Lưu các ảnh đã biến đổi vào thư mục output
        for i, img in enumerate(images_aug):
        # Chuyển đổi từ RGB sang BGR để lưu ảnh bằng OpenCV
            img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
            cv2.imwrite(os.path.join(output_folder, f'{os.path.splitext(image_file)[0]}_augmented_{i+1}.jpg'), img_bgr)
```

```python
# Tạo thư mục 'data' nếu chưa tồn tại
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for thu_muc_con in os.listdir(input_folder):
    duong_dan_thu_muc_con = os.path.join(input_folder, thu_muc_con)
    # Lấy danh sách tất cả các file ảnh trong thư mục input
    image_files = [f for f in os.listdir(duong_dan_thu_muc_con) if os.path.isfile(os.path.join(duong_dan_thu_muc_con, f))]

    # Duyệt qua từng file ảnh trong thư mục input
    for image_file in image_files:
        # Đọc ảnh
        image_path = os.path.join(duong_dan_thu_muc_con, image_file)
        image = cv2.imread(image_path)
        if image is not None:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Tạo các ảnh đã biến đổi
            images_aug = seq.augment_images([image] * 3)

        # Lưu các ảnh đã biến đổi vào thư mục output
            for i, img in enumerate(images_aug):
            # Chuyển đổi từ RGB sang BGR để lưu ảnh bằng OpenCV
                img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
                cv2.imwrite(os.path.join(output_folder, f'{os.path.splitext(image_file)[0]}_augmented_{i+1}.jpg'), img_bgr)
```
